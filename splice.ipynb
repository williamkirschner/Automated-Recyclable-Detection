```python
# Various Prerequisities
import os
import cv2
```

```python
video = cv2.VideoCapture('natural-hdpe.MP4')

success, image = video.read()
count = 0

while success: # while next frame present
  if count <= 7560 and count % 10 == 0: # 0-7559, 126 sec (~80%)
    cv2.imwrite(os.path.join("footage/", "frame%d.jpg" % count), image)
    success, image = video.read()
    if count % 60 == 0:
      print("Train: Converted", count//60, "seconds of footage.")
    count+=1
  elif count > 7560 and count < 9390 and count % 10 == 0: # 7559-, 30.5 sec (~20%)
    cv2.imwrite(os.path.join("footage/", "frame%d.jpg" % count), image)
    success, image = video.read()
    if count % 60 == 0:
      print("Validation: Converted", count//60, "seconds of footage.")
    count+=1
  else:
    success, image = video.read()
    count+=1
print(count)
video.release()
cv2.destroyAllWindows()
```
